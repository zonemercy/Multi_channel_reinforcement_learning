```python
from pymatbridge import Matlab
import time
import numpy as np
import pandas as pd
import random
import gc
import os

PATH = os.path.expanduser("~") + "/Dropbox/MC_RL/matlab/mc_gym/Copy_of_oraginal.m"

# mlab = Matlab(matlab='/Applications/MATLAB_R2014a.app/bin/matlab')
mlab = Matlab(matlab='/usr/local/MATLAB/R2016b/bin/matlab')
mlab.stop()
mlab.start()

l_lam=[]
l_d=[]
l_n0=[]
l_nn=[]
l_esb=[]
l_pd=[]

for i in range(5000):

    LAM = round(random.uniform(0.1, 20),3)
    D = round(np.random.uniform(5,40),2)
    N_0 = round(random.uniform(5, 20),3)
    N_n = np.random.randint(5,20)
      
#     print 'Step: {}; Lambda: {}; Distance: {}; N0:{}; Number of Nodes: {}'\
#             .format(i, LAM, D, N_0, N_n)

    res = mlab.run(PATH, {'arg1': 1, 'arg2': LAM, 'arg3': D, 'arg4':N_0, 'arg5':N_n})
#     print res['result']
    esb = res['result'][:4]
#     n_c = np.where(esb == np.min(esb))[0][0]
    p_id = res['result'][4:]
    
    l_lam.append(LAM)
    l_d.append(D)
    l_n0.append(N_0)
    l_nn.append(N_n)
    l_esb.append(esb)
    l_pd.append(p_id)
    
    if i%500==0:
        print i

df = pd.DataFrame({ 'LAM':l_lam,'D':l_d,'N_0':l_n0,'N_n':l_nn,'esb':l_esb, 'p_id':l_pd },
                  columns=['LAM', 'D', 'N_0', 'N_n', 'esb','p_id'])
df
```
