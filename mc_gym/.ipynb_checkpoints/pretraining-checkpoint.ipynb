```python
from __future__ import division
import numpy as np
import pandas as pd
import xgboost as xgb
from xgboost import plot_tree
import sklearn
from sklearn import model_selection, preprocessing, ensemble
from sklearn.metrics import log_loss, accuracy_score
import itertools
import matplotlib.gridspec as gridspec
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from mlxtend.classifier import EnsembleVoteClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
# from mlxtend.data import iris_data
from mlxtend.plotting import plot_decision_regions
from sklearn import neighbors
from mlxtend.classifier import StackingClassifier, StackingCVClassifier
from mlxtend.classifier import SoftmaxRegression
import matplotlib.pyplot as plt
# %matplotlib inline
# import seaborn as sns
# sns.set_style('darkgrid')
save_file_path = '/Users/zonemercy/Desktop/LATEX/dissertation_2017/images/c5/'
```

```python
df = pd.read_csv('train12000.csv')
df['single_energy'] = df['esb'].apply(lambda x : map(float, x[1:-1].split(", "))[0])
df['esb'] = df['esb'].apply(lambda x : map(float, x[1:-1].split(", ")))
df['result'] = df['esb'].apply(lambda x : np.where(x == np.min(x))[0][0])
df['result_e'] = df['esb'].apply(lambda x : np.min(x))
df=df[['LAM','D','N_0','N_n','result']]
df.head()
```

```python
def add_noise(c=[],v=0.01,data=df):
    for i in c:
        max_value = df[i].max()
        min_value = df[i].min()
        temp = df[i] * np.random.normal(1,v,len(df[i]))
        temp = np.fmin(temp,np.repeat(max_value,len(df[i])))
        temp = np.fmax(temp,np.repeat(min_value,len(df[i])))
        df[i] = temp
        print max_value, min_value
add_noise(c=['LAM','D','N_0'], v=0.02, data=df)
```

```python
param = {}
param['objective'] = 'multi:softprob'
param['eta'] = 0.1
param['max_depth'] = 4
param['silent'] = 1
param['num_class'] = 4
param['eval_metric'] = "mlogloss"
param['min_child_weight'] = 1
# param['subsample'] = 0.7
# param['colsample_bytree'] = 0.7
plst = list(param.items())
print plst
                      
train_X = df[['LAM','D','N_0','N_n']]
train_X=train_X.rename(columns = {'LAM':'Phi'})
train_y = df['result']
                      
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=2016)
for dev_index, val_index in kf.split(range(train_X.shape[0])):
    dev_X, val_X = train_X.loc[dev_index,:], train_X.loc[val_index,:]
    dev_y, val_y = train_y[dev_index], train_y[val_index]
    print dev_X.shape, val_X.shape
    break
```

```python
param['objective'] = 'multi:softprob'
param['max_depth'] = 5

xgtrain = xgb.DMatrix(dev_X, label=dev_y)
xgtest = xgb.DMatrix(val_X, label=val_y)

num_rounds=200
watchlist = [ (xgtrain,'train'), (xgtest, 'test') ]
model = xgb.train(param, xgtrain, num_rounds, watchlist,verbose_eval=20) 
pred_test_y = model.predict(xgtest)
print log_loss(val_y, pred_test_y)
```

```python
%matplotlib inline
plt.figure(figsize=(12, 9))
plot_tree(model,num_trees=10)
# plt.savefig(save_file_path+'c5-tree.pdf')
# plt.show()
```

```python
param['objective'] = 'multi:softmax'

xgtrain = xgb.DMatrix(dev_X, label=dev_y)
xgtest = xgb.DMatrix(val_X, label=val_y)

num_rounds=2000
watchlist = [ (xgtrain,'train'), (xgtest, 'test') ]
model = xgb.train(plst, xgtrain, num_rounds, watchlist,verbose_eval=20) 
pred_test_y = model.predict(xgtest)
print accuracy_score(val_y, pred_test_y)
```
